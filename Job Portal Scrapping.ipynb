```python
import requests
import pandas as pd
from bs4 import BeautifulSoup
import string
```

```python
url2 = "https://www.placementindia.com/job-search/search.php?seeker_search_keyword=Data+Analyst&seeker_search_city=India&seeker_search_experience="
p2 = requests.get(url2)
print(p2.status_code)
```

```python
soup2 = BeautifulSoup(p2.content,'html.parser')
```

```python
jobs = soup2.find_all('div',class_ = 'sjci')
jobs
```

```python
len(jobs)
```

```python
df = pd.DataFrame(columns = ['Job_role','Company_name','Job_link'])
index = 0
for i in jobs:
    try:
        jj = soup2.find_all('a',class_='job-name')
        link = jj['href']
        name = jj.text
        emp = soup2.find_all('p',class_ = 'job-cname')
        comp = emp.text
        df.loc[index,'Job_role'] = name
        df.loc[index,'Company_name'] = comp
        df.loc[index,'Job_link'] = link
    except:
        continue
    
```
